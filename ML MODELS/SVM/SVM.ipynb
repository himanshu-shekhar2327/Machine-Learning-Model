```python
Q1 = data["sepal width (cm)"].quantile(0.25)
Q3 = data["sepal width (cm)"].quantile(0.75)
IQR = Q3 - Q1

outliers = data[(data["sepal width (cm)"] < Q1 - 1.5*IQR) | (data["sepal width (cm)"] > Q3 + 1.5*IQR)]
print(outliers)
```

```python
# implementing the svm model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

X = data[['petal length (cm)','petal width (cm)']].values  # petal length, petal width
y = (data["species"] == 2).astype(int)
  # Iris-Virginica
```

```python
# X
```

```python
y.unique()
```

```python
svm_clf = Pipeline([
 ("scaler", StandardScaler()),
 ("linear_svc", LinearSVC(C=1, loss="hinge")),
 ])
svm_clf.fit(X, y)
```

```python
svm_clf.predict([[5.5, 1.7]])
```
