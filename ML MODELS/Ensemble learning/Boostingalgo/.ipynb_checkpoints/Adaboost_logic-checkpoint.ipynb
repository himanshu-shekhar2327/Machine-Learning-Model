```python
# implementing the Adaboost
import pandas as pd
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.DataFrame()
```

```python
df['X1'] = [1,2,3,4,5,6,6,7,9,9]
df['X2'] = [5,3,6,8,1,9,5,8,9,2]
df['label'] = [1,1,0,1,0,1,0,1,0,0]
```

```python
df.shape
```

```python
df
```

```python
# Plotting the datapoints
import seaborn as sns
sns.scatterplot(x=df['X1'],y=df['X2'],hue=df['label'])
```

```python
df['weight'] = 1/df.shape[0]
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt1 = DecisionTreeClassifier(max_depth=1)
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,2].values
```

```python
#step 2 - Train 1st model
dt1.fit(X,y)
```

```python
#This is the Decision Stump
from sklearn.tree import plot_tree
plot_tree(dt1)
```

```python
# Decision Boundary
plot_decision_regions(X,y,clf=dt1,legend=2)
```

```python
# Finding the prediciton from the decision Stump 1
df['y_pred'] = dt1.predict(X)
```

```python
df
```

```python
# Now Calculating the Weight of both misclassified point  and classified point by
# Using the alpha1 (weight of decision stump 1) 
def calculate_model_weight(error):
    return 0.5 * np.log((1-error)/(error+0.0000001))
```

```python
# Step 3 - Calculate the model weights
alpha1 = calculate_model_weight(0.3)
# Here error is the sum of weight of misclassified Point
alpha1
```

```python
# Step 4 Update the weights so that tell the next model about this
def update_row_weight(row,alpha=0.423):
    if row['label'] == row['y_pred']:
        return row['weight'] * np.exp(-alpha)
    else:
        return row['weight'] * np.exp(alpha)
```

```python
df['updated_weight'] = df.apply(update_row_weight,axis=1)
```

```python
df
```

```python
df['updated_weight'].sum()
```

```python
df['normalized_weight'] = df['updated_weight'] / df['updated_weight'].sum()
```

```python
df
```

```python
df['normalized_weight'].sum()
```

```python
# Creating the range
df['cumsum_upper'] = np.cumsum(df['normalized_weight'])
```

```python
df['cumsum_lower'] = df['cumsum_upper'] - df['normalized_weight']
```

```python
df[['X1','X2','label','weight','y_pred','updated_weight','cumsum_lower','cumsum_upper']]
```

```python
def create_new_dataset(df):
    indices =[]

    for i in range(df.shape[0]):
        a = np.random.random()
        for index,row in df.iterrows():
            if row['cumsum_upper'] > a and a > row['cumsum_lower'] :
                indices.append(index)
    return indices
```

```python
indices_values = create_new_dataset(df)
```

```python
indices_values # This step is called the Upsampling
```

```python
second_df = df.iloc[indices_values,[0,1,2,3]]
```

```python
second_df
```

```python
dt2 = DecisionTreeClassifier(max_depth=1)
X = second_df.iloc[:,0:2].values
y = second_df.iloc[:,2].values
```

```python
dt2.fit(X,y)
```

```python
plot_tree(dt2)
```

```python
plot_decision_regions(X,y,clf=dt2,legend=2)
```

```python
second_df['y_pred'] = dt2.predict(X)
```

```python
second_df
```

```python
alpha2=calculate_model_weight(0.2)
```

```python
alpha2
```

```python
#step 4 Updated Weights
def update_row_weight(row,alpha=0.693):
    if row['label'] == row['y_pred'] :
        return row['weight'] * np.exp(-alpha)
    else:
        return row['weight'] * np.exp(alpha)
```

```python
second_df['updated_weight'] = second_df.apply(update_row_weight,axis=1)
```

```python
second_df
```

```python
second_df['normalized_weight'] = second_df['updated_weight'] / second_df['updated_weight'].sum()
```

```python
second_df[['X1','X2','label','weight','y_pred','normalized_weight','cumsum_lower','cumsum_upper']]
```
