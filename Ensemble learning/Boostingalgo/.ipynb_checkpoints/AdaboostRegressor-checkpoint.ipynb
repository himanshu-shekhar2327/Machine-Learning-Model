```python
# Implementing the Adaboost Regressor
# We first create a dataframe for our data.
import pandas as pd
data = {
    'age' : [23, 31, 35, 35, 42, 43, 45, 46, 46, 51],
    'like goats' : [0, 1, 0, 0, 0, 1, 1, 1, 0, 1], 
    'like height' : [0, 1, 1, 0, 0, 1, 0, 1, 1, 1],
    'climbed meters' : [200, 700, 600, 300, 200, 700, 300, 700, 600, 700]
}
df = pd.DataFrame(data)
df.head()
```

```python
from sklearn.ensemble import AdaBoostRegressor
X = df[['age' ,'like goats' ,'like height']].values
y = df[['climbed meters']].values.reshape(-1,)
reg = AdaBoostRegressor(n_estimators=3,random_state=42)
reg.fit(X,y)
```

```python
# We can then make prediction using the predict method and print the score , which is defined as the mean accuracy
y_hat = reg.predict(X)
print(f"predicition :{y_hat}")
print(f"Score :{reg.score(X,y)}")
```

- This leads to the prediciton [250. 700. 600. 300. 200. 700. 300. 700. 700. 700.] 
- and a score pf 0.97 . Additionally , we can also print the predictions of the individual 
-  model for the three stages using the method stagesd_predict

```python

staged_predictions = [ p for p in reg.staged_predict(X)]
staged_predictions
```

- which shows that all three models yield to different predictions for some samples. The influences are called estimator_weights_ in sklearn and can also be printed.

```python
reg.estimator_weights_
```

- For this example the weights are [0.84729786, 2.07731083, 1.53214395]
- These weights are used for the final prediction, which is achieved by calculating the weighted mean of the individual predictions, with the weights being the influences of the underlying models. Let’s consider a concrete example and make predictions for one sample of the dataset.
